```python
import pandas as pd
import numpy as np
import csv
```

```python
data = pd.read_csv('data/634629.csv')
```

```python
data.head(3)
```

```python
data = data[data.SNWD >= 0]
```

```python
data['datum'] = data.DATE.apply(lambda x: str(x)[0:4] + '-' + str(x)[4:6] + '-' + str(x)[6:8])
```

```python
data.datum = pd.to_datetime(data.datum)
```

```python
with open('data/first_snow.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=';', quotechar='|', quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(['date', 'station', 'snowdepth_mm'])
    for year in range(1970, 2015):
        dat = data[(data.datum >= str(year) + '-09-15') & (data.datum <= str(year + 1) + '-03-31')]
        if (len(dat) > 0):
            dat.sort('datum', inplace=True)
            spamwriter.writerow([str(dat.datum.iloc[0]), str(dat.STATION_NAME.iloc[0]), str(dat.SNWD.iloc[0])])
        else:
            continue
```

```python
data.sort('datum', inplace=True)
```
